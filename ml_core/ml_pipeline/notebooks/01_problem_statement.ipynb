## 1. Problem Statement

### 1.1 The Stock Market Prediction Challenge

Stock market prediction is one of the most challenging problems in financial analytics. Investors, traders, and financial institutions constantly seek methods to anticipate market movements to make informed decisions. The ability to predict stock price movements can lead to:

- **Better investment decisions**
- **Improved portfolio management**
- **Risk mitigation strategies**
- **Automated trading systems**

### 1.2 Why Predicting Exact Price is Difficult

Predicting the exact future stock price is extremely challenging due to several factors:

| Factor | Description |
|--------|-------------|
| **Market Volatility** | Stock prices fluctuate rapidly due to supply-demand dynamics |
| **Random Walk Theory** | Price movements often follow unpredictable patterns |
| **External Events** | Geopolitical events, news, and sentiment cause sudden shifts |
| **Information Asymmetry** | Not all market participants have equal access to information |
| **Non-Linear Relationships** | Complex interactions between multiple variables |
| **High Dimensionality** | Numerous factors influence price simultaneously |

The Efficient Market Hypothesis (EMH) suggests that stock prices already reflect all available information, making consistent prediction nearly impossible. Additionally, financial markets exhibit:

- **Non-stationarity**: Statistical properties change over time
- **Heteroscedasticity**: Variance of returns changes with time
- **Fat tails**: Extreme events occur more frequently than normal distribution suggests

### 1.3 Why Trend Classification Over Regression

Given the difficulty in predicting exact prices, this project adopts a **trend classification approach** instead of regression:

| Approach | Target | Complexity | Practical Use |
|----------|--------|------------|---------------|
| **Regression** | Exact price value | Very High | Limited actionability |
| **Classification** | Up or Down trend | Moderate | Directly actionable |

**Advantages of Trend Classification:**

1. **Actionable Signals**: Binary signals (Buy/Sell) are directly usable for trading
2. **Noise Reduction**: Small price fluctuations are filtered out
3. **Clearer Evaluation**: Accuracy, precision, and recall provide interpretable metrics
4. **Robustness**: Less sensitive to outliers compared to regression
5. **Risk Management**: Easier to set stop-loss and take-profit levels

## 5. Machine Learning Problem Formulation

### 5.1 Problem Type

This project is formulated as a **Binary Classification Problem**.

### 5.2 Target Variable Definition

The target variable represents the next-day price movement direction:

**Target Rule:**
- **Target = 1** if Close(t+1) > Close(t) → Uptrend
- **Target = 0** if Close(t+1) ≤ Close(t) → Downtrend

````python
# Target Variable Formula (Illustrative)
def calculate_target(close_today, close_tomorrow):
    """
    Calculate binary target for trend prediction.
    
    Args:
        close_today: Today's closing price
        close_tomorrow: Tomorrow's closing price
    
    Returns:
        1 if uptrend, 0 if downtrend
    """
    if close_tomorrow > close_today:
        return 1  # Uptrend
    else:
        return 0  # Downtrend
````

### 5.3 Classification Framework

```
┌────────────────────────────────────────────────────────┐
│              BINARY CLASSIFICATION SETUP               │
├────────────────────────────────────────────────────────┤
│                                                        │
│   INPUT FEATURES                    TARGET             │
│   ┌────────────────┐               ┌─────────┐        │
│   │ Price Data     │               │         │        │
│   │ Technical Ind. │    ────►      │  0 or 1 │        │
│   │ Global Markets │    MODEL      │         │        │
│   │ News Sentiment │    ────►      │  Down   │        │
│   │ GDELT Events   │               │   or    │        │

│   └────────────────┘               └─────────┘        │

```

### 5.4 Important Considerations

| Aspect | Implementation |
|--------|----------------|
| **Data Leakage Prevention** | Target is shifted by 1 day to avoid look-ahead bias |
| **Time Series Split** | No random splitting; chronological train/validation/test |
| **Feature Lagging** | Global and news data lagged by 1 day |
| **Class Balance** | Monitor for class imbalance in target variable |

## 6. Volatility-Based Stock Classification

### 6.1 Volatility Calculation

Stock volatility is measured using the **standard deviation of daily returns**.

**Mathematical Formula:**

```
Daily Return = (Close_t - Close_t-1) / Close_t-1

Volatility = Standard Deviation of Daily Returns
```

````python
# Volatility Calculation (Illustrative)
import numpy as np

def calculate_volatility(prices):
    """
    Calculate stock volatility as standard deviation of daily returns.
    
    Args:
        prices: Array of closing prices
    
    Returns:
        Volatility (standard deviation of returns)
    """
    returns = np.diff(prices) / prices[:-1]
    volatility = np.std(returns)
    return volatility
````

### 6.2 Volatility Categories

Stocks are classified into three categories based on their volatility percentile:

```
┌────────────────────────────────────────────────────────────────┐
│                    VOLATILITY DISTRIBUTION                      │
├────────────────────────────────────────────────────────────────┤
│                                                                 │
│     ▲                                                          │
│     │                         ┌──────────┐                     │
│     │                         │ VOLATILE │                     │
│  F  │            ┌────────────┤ Top 33%  │                     │
│  r  │            │  STABLE    └──────────┘                     │
│  e  │  ┌─────────┤ Middle 33% │                                │
│  q  │  │NON-VOL. └────────────┘                                │
│  u  │  │Bottom   │                                             │
│  e  │  │  33%    │                                             │
│  n  │  └─────────┘                                             │
│  c  │                                                          │
│  y  └──────────────────────────────────────────────────►       │
│                         Volatility (σ)                         │
│                                                                 │
└────────────────────────────────────────────────────────────────┘
```

### 6.3 Category Definitions

| Category | Percentile Range | Characteristics | Example Sectors |
|----------|-----------------|-----------------|------------------|
| **Volatile** | Top 33% (67th - 100th) | High price swings, higher risk/reward | Small-cap, Tech, Pharma |
| **Stable** | Middle 33% (33rd - 67th) | Moderate movements | Mid-cap, Banking |
| **Non-Volatile** | Bottom 33% (0 - 33rd) | Low price swings, steady growth | Large-cap, FMCG, Utilities |

````python
# Volatility Classification Logic (Illustrative)
import numpy as np

def classify_volatility(volatility_values):
    """
    Classify stocks into volatility categories.
    
    Args:
        volatility_values: Array of volatility values for all stocks
    
    Returns:
        List of category labels
    """
    p33 = np.percentile(volatility_values, 33)
    p67 = np.percentile(volatility_values, 67)
    
    categories = []
    for v in volatility_values:
        if v <= p33:
            categories.append('Non-Volatile')
        elif v <= p67:
            categories.append('Stable')
        else:
            categories.append('Volatile')
    
    return categories
````

## 8. Evaluation Metrics

### 8.1 Metrics Overview

Model performance is evaluated using multiple classification metrics:

```
┌────────────────────────────────────────────────────────────────┐
│                    EVALUATION METRICS                          │
├────────────────────────────────────────────────────────────────┤
│                                                                 │
│   ┌──────────┐  ┌───────────┐  ┌────────┐  ┌──────────┐       │
│   │ ACCURACY │  │ PRECISION │  │ RECALL │  │ F1 SCORE │       │
│   └──────────┘  └───────────┘  └────────┘  └──────────┘       │
│                                                                 │
│                  ┌───────────────────┐                         │
│                  │ CONFUSION MATRIX  │                         │
│                  └───────────────────┘                         │
│                                                                 │
└────────────────────────────────────────────────────────────────┘
```

### 8.2 Metric Definitions

#### Accuracy
```
Accuracy = (TP + TN) / (TP + TN + FP + FN)
```
**Interpretation:** Overall correctness of predictions.

#### Precision
```
Precision = TP / (TP + FP)
```
**Interpretation:** Of all predicted uptrends, how many were actually uptrends?

#### Recall (Sensitivity)
```
Recall = TP / (TP + FN)
```
**Interpretation:** Of all actual uptrends, how many did we correctly predict?

#### F1 Score
```
F1 = 2 × (Precision × Recall) / (Precision + Recall)
```
**Interpretation:** Harmonic mean of precision and recall.

### 8.3 Confusion Matrix

```
                      PREDICTED
                 ┌─────────┬─────────┐
                 │ Down(0) │  Up(1)  │
        ┌────────┼─────────┼─────────┤
        │Down(0) │   TN    │   FP    │
ACTUAL  ├────────┼─────────┼─────────┤
        │ Up(1)  │   FN    │   TP    │
        └────────┴─────────┴─────────┘
        
Where:
  TP = True Positive  (Correctly predicted Up)
  TN = True Negative  (Correctly predicted Down)
  FP = False Positive (Incorrectly predicted Up)
  FN = False Negative (Incorrectly predicted Down)
```
